```python
#Loading dependencies
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading dataset
df=pd.read_csv('./titanic (1).csv')
df.head()
```

```python
#dropping columns
df.drop(columns=['PassengerId','Name','Cabin','Ticket'],inplace=True)
```

```python
#dropping null values and duplicates
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

#### groupby

```python
#gender wise survival
sex_survived_sum = df.groupby(['Sex'])['Survived'].sum()
```

```python
sex_survived_sum
```

```python
#gender wise survival,average
sex_survived_average=df.groupby(['Sex'])['Survived'].mean()*100
```

```python
sex_survived_average
```
